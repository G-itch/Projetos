```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
arr = np.linspace(0,5,11)
```

```python
arr
```

```python
y = arr ** 2
```

```python
y
```

```python
plt.plot(arr,y,"b--")
plt.xlabel("Eixo X")
plt.ylabel("Eixo Y")
plt.title("Título")
```

```python
plt.subplot(1,2,1)
plt.plot(arr,y,"r--")
plt.subplot(1,2,2)
plt.plot(y,arr,"g*-")
```

```python
fig = plt.figure()
axes1= fig.add_axes([0.1,0.1,0.8,0.8])
axes2= fig.add_axes([0.2,0.5,0.3,0.3])
axes2.plot(y,arr)
axes2.set_xlabel("Eixo Y")
axes2.set_title("Inverso")
axes1.plot(arr,y)
axes1.set_xlabel("Eixo X")
axes1.set_title("Título")
```

```python
fig, ax = plt.subplots()
ax.plot(arr,arr**3)
ax.plot(arr,arr**4,"r-.")
ax.set_xlabel("Eixo X")
ax.set_ylabel("Eixo Y")
ax.set_title("Título")
```

```python
fig, ax = plt.subplots(nrows=1,ncols=2)
```

```python
ax[0].plot(arr,y,"b")
ax[0].set_title("Título1")

ax[1].plot(y,arr,"y--")
ax[1].set_title("Título2")
fig
```

```python
fig, ax = plt.subplots(5,5)
plt.tight_layout()
```

```python
fig = plt.figure(figsize=(8,4),dpi=100)
```

```python
fig, axes = plt.subplots(figsize=(8,8))
axes.plot(arr,y,"g*-",label="X ** 2")
axes.plot(y,arr,"y--",label="X ** 0.5")
axes.set_title("Títúlo")
axes.legend()
```
