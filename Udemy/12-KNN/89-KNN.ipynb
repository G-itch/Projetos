```python
import numpy as np 
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd 
%matplotlib inline
```

```python
df = pd.read_csv("Classified Data",index_col = 0)
```

```python
df
```

```python
df.columns
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(df.drop("TARGET CLASS",axis=1))
```

```python
df_p = scaler.transform(df.drop("TARGET CLASS",axis=1))
```

```python
df_p
```

```python
df_param = pd.DataFrame(df_p,columns = df.columns[:-1])
```

```python
df_param
```

```python
df.columns[:-1]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test= train_test_split(df_param,df["TARGET CLASS"],test_size=0.3)
```

```python
x_train
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors=1)
```

```python
knn.fit(x_train,y_train)
```

```python
predict = knn.predict(x_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
classification_report(y_test,predict)
```

```python
print(confusion_matrix(y_test,predict))
```

```python
error_rate=[]
for i in range(1,40):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(x_train,y_train)
    predict = knn.predict(x_test)
    error_rate.append(np.mean(predict != y_test))
```

```python
error_rate
```

```python
min1 = error_rate[0]
for i in error_rate:
    if i < min1:
        min1 = i
print(min1,)
```

```python
plt.figure(figsize=(12,6))
plt.style.use("seaborn")
plt.plot(range(1,40),error_rate,"b--",marker="o")
plt.xlabel("K numbers")
plt.ylabel("Error")
plt.title("Error rate of K variation")
```

```python
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(x_train,y_train)
predict = knn.predict(x_test)
print(classification_report(y_test, predict))
```
