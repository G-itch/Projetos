```python
import numpy as np 
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
cancer = load_breast_cancer()
```

```python
cancer["feature_names"]
cancerSVM = pd.DataFrame(cancer["data"],columns=cancer["feature_names"])
```

```python
cancerSVM
```

```python
cancer.keys()
```

```python
cancer["target_names"]
```

```python
cancer["feature_names"]
```

```python
plt.figure(dpi=300,figsize=(20,20))
sns.heatmap(cancerSVM.corr(),annot=True,fmt=".2f")
```

```python
cancer_target = pd.DataFrame(cancer["target"],columns=["Cancer"])
```

```python
cancer_target
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(cancerSVM,cancer_target,test_size=0.3,random_state=101)
```

```python
x_train
```

```python
y_train
```

```python
from sklearn.svm import SVC
```

```python
svm = SVC()
```

```python
svm.fit(x_train,y_train)
```

```python
pred = svm.predict(x_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(y_test,pred))
```

```python
print(confusion_matrix(y_test,pred))
```

```python
p_grid = {"C":[0.1,1,10,100,1000],"gamma":[1,0.1,0.01,0.001,0.0001],"kernel":["rbf"]}


from sklearn.model_selection import GridSearchCV
```

```python

grid = GridSearchCV(SVC(),p_grid,refit=True,verbose=3)
grid.fit(x_train,y_train)
```

```python
grid.best_params_
```

```python
pred = grid.predict(x_test)
```

```python
print(classification_report(y_test,pred))
print(confusion_matrix(y_test,pred))
```
