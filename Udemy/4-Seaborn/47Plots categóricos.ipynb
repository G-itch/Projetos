```python
import seaborn as sns
```

```python
import numpy as np
```

```python
%matplotlib inline
```

```python
tips = sns.load_dataset("tips")
```

```python
tips.head()
```

```python
sns.barplot("sex","total_bill",data=tips)
```

```python
sns.barplot("sex","total_bill",data=tips,estimator=np.std)
```

```python
sns.countplot("sex",data=tips)
```

```python
sns.boxplot("day","total_bill",data=tips,palette="Blues",hue="sex")
```

```python
sns.boxplot(data=tips,orient="h")
```

```python
sns.violinplot("day","tip",data=tips,hue="sex",split=True)
```

```python
sns.stripplot("day","total_bill",data=tips,hue="sex",split=True)
```

```python
sns.swarmplot("day","total_bill",data=tips,hue="sex",split=True)
```

```python
sns.swarmplot("day","total_bill",data=tips,color="black")
sns.violinplot("day","total_bill",data=tips)
```

```python
sns.factorplot("sex","total_bill",data=tips)
```
