```python
import seaborn as sns
%matplotlib inline
```

```python
tips = sns.load_dataset("tips")
```

```python
tips.head()
```

```python
tips.info()
```

```python
sns.lmplot("total_bill","tip",data=tips,hue="sex",palette="magma",markers=["o","v"],scatter_kws={"s":100})
```

```python
sns.lmplot("total_bill","tip",data=tips,hue="sex",palette="cool",col="sex",row="time",aspect=0.6)
```
