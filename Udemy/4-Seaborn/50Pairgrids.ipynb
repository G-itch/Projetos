```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
iris= sns.load_dataset("iris")
```

```python
iris["species"].value_counts()
```

```python
g = sns.PairGrid(iris)
g.map_diag(plt.hist)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```

```python
sns.pairplot(iris,hue="species")
```
