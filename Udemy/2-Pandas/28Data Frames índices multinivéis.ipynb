```python
lista1 = ["Id","Nome","Número","Anos","Social","A"]
lista2 = ["U","V","W","X","Y","Z"]
lp = lista1 + lista2
x = input("O que deseja? ")
plv = x.split()
pl = []
pl += ([v for v in plv if v in lista1])
plp = []
plp += ([v for v in plv if v in lp])
for x in pl:
    plv.remove(x)
pl2 = []
pl2 += ([v for v in plv if v in lista2])
```
