```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.style.use("seaborn")
%matplotlib inline
```

```python
import pandas as pd 
```

```python
import numpy as np
```

```python
from sklearn.datasets import make_blobs
```

```python
data = make_blobs(n_samples=200,n_features=2,centers=4,cluster_std=1.8,random_state=50)
```

```python
data[1]
```

```python
plt.figure(figsize=(12,10))
plt.scatter(data[0][:,0],data[0][:,1],c=data[1],cmap="cool")
```

```python
from sklearn.cluster import KMeans
```

```python
K_means = KMeans(n_clusters=2)
```

```python
K_means.fit(data[0])
```

```python
K_means.cluster_centers_ 
```

```python
K_means.labels_
```

```python
f, (ax1,ax2) = plt.subplots(1,2,sharey=True,figsize=(12,8))
ax1.set_title("K Means")
ax1.scatter(data[0][:,0],data[0][:,1],c=K_means.labels_,cmap="cool")
ax2.set_title("Original")
ax2.scatter(data[0][:,0],data[0][:,1],c=data[1],cmap="cool")
```
